<samp>
- Age: Idade do paciente <br/>
- Sex: Sexo (M/F) <br/>
- ChestPainType: Tipo de dor no peito (ATA, NAP, ASY, etc.) <br/>
- RestingBP: Pressão arterial em repouso <br/>
- Cholesterol: Nível de colesterol <br/>
- FastingBS: Nível de açúcar no sangue em jejum (0 ou 1) <br/>
- RestingECG: Resultados do eletrocardiograma em repouso <br/>
&nbsp&nbsp -- Normal: Normal <br/>
&nbsp&nbsp -- ST: Apresentando anormalidade na onda ST-T <br/>
&nbsp&nbsp -- LVH: Indica hipertrofia ventricular esquerda <br/>
- MaxHR: Frequência cardíaca máxima atingida <br/>
- ExerciseAngina: Angina induzida por exercício (Y/N) <br/>
- Oldpeak: Depressão do segmento ST <br/>
- ST_Slope: Inclinação do segmento ST <br/>
- HeartDisease: Diagnóstico de doença cardíaca (0 = não, 1 = sim) <br/>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar a base de dados
df = pd.read_csv('heart.csv')
display(df.head())

# Verificar os tipos de dados
display(df.info())

# Verificar se há dados faltantes
display(df.isnull().sum())

# Gerar uma estatística descritiva
display(df.describe())
```

```python
# Distribuição por idade
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Distribuição por idade')
plt.show()
```

```python
# Contagem de pacientes com doença cardiaca
sns.countplot(x='HeartDisease', data=df)
plt.title('Contagem HeartDisease')
plt.show()
```

```python
# Boxplot por idade x doença
sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Idade x Doença')
plt.show()
```

```python
# Relação MaxHR com Doença
sns.boxplot(x='HeartDisease', y='MaxHR', data=df)
plt.title('MaxHR x Doença')
plt.show()
```

```python
df.columns
```

```python
# Contagem por tipo de dores no peito (ChestPainType)
sns.countplot(x='ChestPainType', data=df, order=df['ChestPainType'].value_counts().index)
plt.title('Tipos de Dores no Peito')
plt.show()
```

```python
# Avaliar a Distribuição do Colesterol
sns.histplot(df['Cholesterol'], bins = 20, kde=True, color='green')
plt.title('Distribuição do Colesterol')
plt.show()
```

```python
# Matriz de Correlação
cm = df.select_dtypes(include=['float64','int64']).corr()
cm
```

```python
sns.heatmap(cm, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```
