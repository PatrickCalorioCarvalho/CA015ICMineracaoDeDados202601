### **1. Normalização Z-Score (Padronização)**
Transforma os dados para que tenham **média 0** e **desvio padrão 1**. A fórmula usada é:

$Z = \frac{X - \mu}{\sigma}$

Onde:
- ($X$) = valor original
- ($\mu$) = média da variável
- ($\sigma$) = desvio padrão da variável


### **2. Normalização Min-Max (Escalonamento)**
Transforma os dados para um intervalo **entre 0 e 1 (ou -1 e 1)**, usando a fórmula:

$
X' = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}
$

Onde:
- ($X$) = valor original
- ($X_{\text{min}}$) = menor valor da variável
- ($X_{\text{max}}$) = maior valor da variável




```python
# Z-score
# remoção das colunas id e stroke
num_cols.remove('id')
num_cols.remove('stroke')

print("Antes:")
display(df[num_cols].head())

for col in num_cols:
  media = df[col].mean()
  desvio = df[col].std()
  df[col] = (df[col] - media)/desvio

print("Depois")
display(df[num_cols].head())
```

#### Transformação das colunas categóricas

```python
cat_cols = df.select_dtypes(include="object").columns.tolist()
print(cat_cols)
```

```python
for cat in cat_cols:
  print(f"{cat}: {df[cat].unique()}")
```

```python
# Map (colunas com até 3 categorias)
df['gender'] = df['gender'].map({'Male':0, 'Female':1, 'Other':2})
df['ever_married'] = df['ever_married'].map({'Yes':1, 'No':0})
df['Residence_type'] = df['Residence_type'].map({'Urban':1, 'Rural':0})
df[['gender','ever_married','Residence_type']].head()
```

```python
# Transformação com get_dummies (hot_encoder)
dummies = pd.get_dummies(df[['work_type','smoking_status']],
                         prefix=['work','smoke'])
dummies
```

```python
# Remover as colunas categóricas originais de 'df'
df = df.drop(columns=['work_type','smoking_status'])
```

```python
df.columns
```

```python
# concatenar as ciolunas dummies no dataset original
df = pd.concat([df,dummies], axis=1)
```
